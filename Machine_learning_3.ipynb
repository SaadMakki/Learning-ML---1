### What make a good feature

a good feature makes it easy to decide between two different things

I am gonna write a classifier to tell about 2 different types to dogs. <br>
we are gonna take two types of dogs and septate them based on their height.

```python
import numpy as np
import matplotlib.pyplot as plt
```

creating data for dogs

```python
grayHounds = 500
labs = 500

gray_height = 28 + 4 * np.random.randn(grayHounds)
labs_height = 24 + 4 * np.random.randn(labs)
```

visualizing the data for dog height

```python
plt.hist(
    [gray_height, labs_height], 
    stacked= True, 
    label=['Gray Height', 'Labs Height'],
    color=['#4ECDC4', '#FF6B6B'],
    edgecolor='white',
    )
plt.xlabel('Height Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Height Distribution Comparison', fontsize=14, pad=20)
plt.legend(title='Groups:', title_fontsize=12)
plt.grid(axis='y', alpha=0.3)  # Add subtle grid lines

plt.show()
```

now as you can see the further we go the height feature is not that useful anymore. <br>
so thats why we need good and practical features to train our classifier<br>
the generale rule of thumb is to think like the classifier and then select the features to train it on
